```python
testX = np.append(np.ones((data.shape[0],1)),data[:,:2],axis=1)
testY = data[:,2:3]
print(testX.shape)
print(testY.shape)
```

```python
def init(t0=0,t1=0,t2=0):
    return np.array([[t0],[t1],[t2]])

def synData(p,s):  # p=parameters s=size of set to be generated
    X1 = np.reshape(np.random.normal(size=s),(s,-1))
    X2 = np.reshape(np.random.normal(size=s),(s,-1))
    X2 = np.append(X1,X2,axis=1)
    X = np.append(np.reshape(np.ones((s,1)),(s,-1)),X2,axis=1)
    noise = np.reshape(np.random.normal(0,sqrt(2),s),(s,-1))
    Y=np.dot(X,p)+noise
    return X,Y,noise

def shuffle(X,Y):
    data = np.append(X,Y,axis=1)
    np.random.shuffle(data)  
    return data[:,0:3],data[:,3]

def cost(data,theta,out):
    return ((np.dot(data,theta)-out)**2)/(2*data.shape[0])

def stochasticDescent(X,Y,theta,alpha,steps,epsilon=1e-4,batch_size=1):
    X,Y = shuffle(X,Y)  #Initial Shuffling
    Y=np.reshape(Y,(Y.shape[0],1))
    converged = False
    i=0
    m=X.shape[0]
    num_batches=int(m/batch_size)
    b=0
    avgCost=0
    prevCost = -1
    while (not converged and i<steps):
        curX = X[b*batch_size:(b+1)*batch_size,:]
        curY = Y[b*batch_size:(b+1)*batch_size,:]
        b=int((b+1)%num_batches)
        
        hyp = np.dot(curX,theta)
        grad = np.reshape(np.sum(((curY-hyp)*(-curX)),axis=0)/(curX.shape[0]),(theta.shape))
        thetaNext = theta-alpha*grad
        avgCost = avgCost+cost(curX,thetaNext,out)
#         print(thetaNext)

        if(i%5000==0):
            avgCost = curCost/5000
            if(prevCost!=-1 and abs(prevCost-avgCost)<=epsilon):
                conv = True
            prevCost = avgCost
            avgCost=0
        
        if(i>steps):
            theta=thetaNext
            conv = True
        else:
            theta = thetaNext
        
        i=i+1
    print(i)
    return theta
```

```python
def error(data,out,p):
    print(data.shape)
    print(p.shape)
    return np.sum((out-np.dot(data,p))**2)/(2*data.shape[0])
```

```python
x,y,n=synData(init(3,1,2),1000)
print(x.shape,y.shape)
```

```python
print((np.sum((testY-np.dot(testX,init(3,1,2)))**2))/(2*data.shape[0]))
print(error(testX,testY,init(3,1,2)))
```

```python
st=time.time()
param=stochasticDescent(x,y,init(0,0,0),0.001,10000,batch_size=1)
# param2=stochasticDescent(x,y,init(0,0,0),0.001,1000000,batch_size=100)
# param3=stochasticDescent(x,y,init(0,0,0),0.001,1000000,batch_size=10000)
# param4=stochasticDescent(x,y,init(0,0,0),0.001,1000000,batch_size=1000000)
print(time.time()-st)
```
